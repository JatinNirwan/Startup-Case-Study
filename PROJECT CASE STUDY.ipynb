```python
#Find Percentage of amount Funded by each Invetment Type
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
file = open("startup_funding.csv")
data = csv.DictReader(file,skipinitialspace=True)
Investment = []
for row in data:
    Investment.append(row['InvestmentType'])
Investment = np.array(Investment)
#Investment = Investment(Investment !='')
for row in range(len(Investment)):
    if Investment[row] =='Crowd funding':
        Investment[row] = 'Crowd Funding'
    if Investment[row] =='PrivateEquity':
        Investment[row] = 'Private Equity'
    if Investment[row] =='SeedFunding':
        Investment[row] = 'Seed Funding'
dic=dict()
for row in Investment:
    dic[row]=dic.get(row,0)+1
x = []
y = []
for i in dic:
    x.append(i)
    y.append(dic[i])
x = np.array(x)
y = np.array(y)
x = x[np.argsort(y)]
y = np.sort(y)
for i in range(len(y)):
    print(x[i],y[i])
explode=[10,0,5,0,0]
plt.pie(y,explode=explode,radius=2,labels=x,autopct = "%.4f %%")
plt.show()
```

```python
#Top Investor who invested maximum No of times

import numpy as np
from collections import Counter
import pandas as pd
import matplotlib.pyplot as plt
import csv
df_start = pd.read_csv('startup_funding.csv', encoding='utf-8')
df_start=df_start.dropna(subset=['InvestorsName'])
d={}
def separateCity(n):   
    n=n.split(',')
    l=[]
    
    for i in n:
        l.append(i.strip())
    for i in l:
        if i in d:
            d[i]=d[i]+1
        else:
            d[i]=1
df_start['InvestorsName']=df_start['InvestorsName'].apply(separateCity)
r=Counter(d)
high = r.most_common(5)  
for i in high: 
    print(i[0],i[1]," ") 
    break
```
